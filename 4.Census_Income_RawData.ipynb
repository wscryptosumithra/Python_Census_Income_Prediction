```python
out_col=[' hours-per-week',' fnlwgt','age',' education-num',' capital-gain',' capital-loss']
```

```python
for col in out_col:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3-Q1
  LW = Q1-1.5*IQR #lower whisker
  UW = Q3 + 1.5*IQR #Upper whisker
  df=df[(df[col]>LW) & (df[col]<= UW)]
```
